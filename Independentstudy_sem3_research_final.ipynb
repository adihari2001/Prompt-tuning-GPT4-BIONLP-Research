```python
def prompt1(original_question):
  instruction = """
  You are a medical expert who is preparing questions for the USMLE-step 2-CK exam.
  """

  prompt0 = "There are 5 difficulty levels for the questions: 1-Very Easy, 2-Easy, 3-Medium, 4-Hard, 5-Very Hard. Why is the above question difficulty level 5- Very Hard? What characteristics make the question difficulty level 5?"
  input_titles = original_question + "\n" + prompt0
  response = openai.ChatCompletion.create(
      model="gpt-4",
      messages=[
          {"role": "system", "content": instruction},
          {"role": "user", "content": input_titles},
      ],
      temperature=0
  )
  #print(response["choices"][0]["message"]["content"])
  #print(type(response["choices"][0]["message"]["content"]))
  return response["choices"][0]["message"]["content"]
```
